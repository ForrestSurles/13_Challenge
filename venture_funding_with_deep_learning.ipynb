## Compile and Evaluate a Binary Classification Model Using a Neural Network

### <font color=OrangeRed>**COMBINED STEP 1**</font>

```python
# Define inputs for the function
# to build and compile the model
number_input_features = len(X_train.iloc[0])
hidden_node_layers = []     # List of lists to hold hidden node layer info

# The first sub-item assigns the number of hidden nodes in the layer
# The second sub-item assigns the activation function for the layer

# Hidden node layer 1
hidden_node_layers.append(
        [
                (number_input_features + 1) // 2,
                'relu'
        ]
)

# Hidden node layer 2
hidden_node_layers.append(
        [
                (hidden_node_layers[0][0] + 1) // 2,
                'relu'
        ]
)

# Activation function for the output layer
output_layer_activation = 'sigmoid'

# NOTE: This function currently supports ONLY ONE output neuron
nn = binary_class_neural_net(
        number_input_features,
        hidden_node_layers,
        output_layer_activation
)
```

```python
# Fit the model using 50 epochs and the training data
fit_model = nn.fit(
    X_train_scaled,
    y_train,
    epochs=50
)
```

### Step 3: Evaluate the model using the test data to determine the model’s loss and accuracy.


```python
# Evaluate the model loss and accuracy metrics using the evaluate method and the test data
model_loss, model_accuracy = nn.evaluate(
    X_test_scaled,
    y_test,
    verbose=2
)

# Display the model loss and accuracy results
print(f'\nModel Results:\n    Loss: {model_loss:%}\n    Accuracy: {model_accuracy:%}')
```
